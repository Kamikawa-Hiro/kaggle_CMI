## 欠損値の確認

```python
# 訓練データの欠損値の数を計算
train_missing = train.isnull().sum()
train_missing.name = "train_missing_count" # シリーズに名前を付ける

# テストデータの欠損値の数を計算
test_missing = test.isnull().sum()
test_missing.name = "test_missing_count" # シリーズに名前を付ける

# 2つのシリーズ（欠損数のカウント結果）を結合してデータフレームを作成
# pd.concatは、インデックス（この場合はカラム名）を基準に結合します
missing_df = pd.concat([train_missing, test_missing], axis=1)

# 訓練データの欠損値の数を計算
train_demo_missing = train_demo.isnull().sum()
train_demo_missing.name = "train_demo_missing_count" # シリーズに名前を付ける

# テストデータの欠損値の数を計算
test_demo_missing = test_demo.isnull().sum()
test_demo_missing.name = "test_demo_missing_count" # シリーズに名前を付ける

# 2つのシリーズ（欠損数のカウント結果）を結合してデータフレームを作成
# pd.concatは、インデックス（この場合はカラム名）を基準に結合します
missing_demo_df = pd.concat([train_demo_missing, test_demo_missing], axis=1)

display(missing_df.T)
display(missing_demo_df.T)
```

## sequence_idが等しいときの`gesture`はすべて等しいのか

```python
uniqueness_check = train.groupby('sequence_id')['gesture'].nunique()

# ユニークな値の数が1より大きいグループ（= gestureが複数種類あるグループ）を探す
inconsistent_sequences = uniqueness_check[uniqueness_check > 1]
if inconsistent_sequences.empty:
    print(f"✅ 検証完了: 全ての'sequence_id'において、'gesture'の値はすべて単一でした。")
```
